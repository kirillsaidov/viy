```python
import matplotlib.pyplot as plt # for plotting
import numpy as np # for transformation

import glob
import pathlib

import torch # PyTorch package
import torchvision # load datasets
import torchvision.transforms as transforms # transform data
import torch.nn as nn # basic building block for neural neteorks
import torch.nn.functional as F # import convolution functions like Relu
import torch.optim as optim # optimzer
from torch.utils.data import DataLoader
```

```python
train_path = '../../data/age/train'
test_path = '../../data/age/test'
device = 'cuda' if torch.cuda.is_available() else 'cpu'

epochs = 12
batch_size = 4
imgw, imgh = 128, 128
```

```python
# python image library of range [0, 1] 
# transform them to tensors of normalized range[-1, 1]
transformer = transforms.Compose([
    transforms.Resize((imgw, imgh)),
    transforms.RandomHorizontalFlip(0.3),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(
        mean = [0.63154647, 0.48489257, 0.41346439],
        std = [0.21639832, 0.19404103, 0.18550038]
    )
])

# get folder names which will be our classes
root = pathlib.Path(train_path)
classes = sorted([i.name.split('/')[-1] for i in root.iterdir()])

# remove folders or files starting with a dot '.'
for i in classes:
    if i.startswith("."):
        classes.remove(i)

# save classes
num_classes = len(classes)
with open("classes.txt", "w") as f:
    for i in range(0, num_classes - 1):
        f.write(classes[i] + ";")
    f.write(classes[num_classes - 1])

print(f'{num_classes} classes found: {classes}')

# create a dataloader to load and transform the data automatically
trainloader = DataLoader(
    torchvision.datasets.ImageFolder(train_path, transform = transformer),
    batch_size = batch_size,
    shuffle = True,
)

testloader = DataLoader(
    torchvision.datasets.ImageFolder(test_path, transform = transformer),
    batch_size = batch_size,
    shuffle = False,
)

# calculating the size of training and testing images
img_train_count = len(glob.glob(train_path + '/**/*.jpg'))
img_test_count = len(glob.glob(test_path + '/**/*.jpg'))
print(f'Images count: train, test = {img_train_count}, {img_test_count}')
```

```python
def imshow(img):
  img = img * 0.5 + 0.5 # unnormalize
  npimg = img.numpy() # convert to numpy objects
  plt.imshow(np.transpose(npimg, (1, 2, 0)))
  plt.show()

# get random training images with iter function
dataiter = iter(trainloader)
images, labels = dataiter.next()

# call function on our images
imshow(torchvision.utils.make_grid(images))

# print the class of the image
print(', '.join('%s' % classes[labels[j]] for j in range(batch_size)))
```

```python
class ConvNet(nn.Module):
    def __init__(self):
        super(ConvNet, self).__init__()
        
        self.conv1 = nn.Conv2d(3, 6, 2)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 2)
        self.fc1 = nn.Linear(imgw * imgh, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 3)
        
#         self.net = nn.Sequential(
#             nn.Conv2d(
#                 in_channels = 3,
#                 out_channels = 12,
#                 kernel_size = 2,
#                 stride = 1,
#                 padding = 1
#             ),
#             nn.ReLU(),
#             nn.MaxPool2d(kernel_size = 2),
            
#             nn.Conv2d(
#                 in_channels = 12,
#                 out_channels = 24,
#                 kernel_size = 2,
#                 stride = 1,
#                 padding = 1
#             ),
#             nn.ReLU(),
#             nn.MaxPool2d(2, 2),
            
#             nn.Linear(int(24 * imgw * imgh / 16), 120),
#             nn.ReLU(),
#             nn.Linear(120, 84),
#             nn.ReLU(),
            
#             nn.Flatten(),
#             nn.Linear(84, num_classes),
#         )
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
#         x = self.net(x)
        return x

net = ConvNet()
print(net)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr = 0.001)
```

```python
for epoch in range(epochs):
    running_loss = 0.0
    for i, data in enumerate(trainloader, 0):
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

#         # print statistics
#         running_loss += loss.item()
#         if i % 2000 == 1999:    # print every 2000 mini-batches
#             print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}')
#             running_loss = 0.0

print('Finished Training')
```
